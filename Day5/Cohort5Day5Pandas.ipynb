```
import pandas as pd

def change_value(x):
  return round(float(x))

data = pd.read_csv('/content/drive/My Drive/Advertising.csv',converters={'TV':change_value,'radio':change_value,'newspaper':change_value,'sales':change_value})
data.head()
data.to_csv('convertedads.csv')
data1 = pd.read_csv('convertedads.csv')
data1.head()
```

```
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Advertising.csv')
import matplotlib.pyplot as plt
plt.plot(df[['TV']],df['sales'],'o')
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
```

```
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)
ax1.plot(df['TV'],df['sales'],'o',color='red')
ax1.set_xlabel('TV')
ax1.set_ylabel('sales')
ax2.plot(df['radio'],df['sales'],'+',color='green')
ax2.set_xlabel('radio')
ax3.plot(df['newspaper'],df['sales'],'*',color='k')
ax3.set_xlabel('newspaper')
plt.show()
fig1 = plt.figure()
ax = fig1.add_subplot(111,projection='3d')
ax.scatter(df[['TV']],df[['radio']],df['sales'])
ax.set_xlabel('TV')
ax.set_ylabel('radio')
ax.set_zlabel('sales')
plt.show()
```

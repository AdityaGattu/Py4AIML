## Pandas

### Data Description:

* The dataset is made up of 1016810 rows and 7 columns. 

* Each line is the data of one student

* The seven columns correspond to 
   - Region Code
   - student id
   - English marks
   - First Language marks
   - Math marks
   - Science marks
   - Social marks
* The marks for the students absent  for the exam are sometimes denoted by 'A', sometimes denoted by ' AA', sometimes denoted by 'AAA' and sometimes denoted by 'G'

* If the student is not marked as absent but no marks in a subject then he is a supplementary candidate.

#### Exercise 1 :

Load the dataset using Pandas and drop all the supplementary candidates from the dataset

```python
import pandas as pd

df = pd.read_csv('Data_Munging_SSLC_data.txt',sep=';',names =  ['Region  Code','student id','English marks','First Language marks','Math marks','Science marks','Social marks'])

# instead of names u can use linke df.columns also.
# df.columns = ['Region  Code','student id','English marks','First Language marks','Math marks','Science marks','Social marks']

df.head()

modified = df.dropna(  subset = ['English marks','First Language marks','Math marks','Science marks','Social marks'])

print(modified)
```

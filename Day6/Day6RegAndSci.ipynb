```
from scipy.integrate import quad

def square_func(x):
  return 3*x*x
result,err = quad(square_func,1,2)
print('The integration is ',result)
print('The error is ',err)
```

```
from scipy.integrate import odeint
import numpy as np
def f(y,t):
  return -2 * y * t 
y0 = 1
t = np.arange(0,2,0.01)
y = odeint(f,y0,t)
import matplotlib.pyplot as plt
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```

```
file = open('headbrain.csv','r')
header = file.readline()
x = np.empty((0,),dtype=int)
y = np.empty((0,),dtype=int)
for line in file:
  fields = line.split(',')
  x = np.append(x,int(fields[2]))
  y = np.append(y,int(fields[3]))
```

```
from scipy import stats
slope,intercept,rvalue,pvalue,stderr = stats.linregress(x,y)
print(slope)
print(intercept)
import matplotlib.pyplot as plt
plt.plot(x,y,'o',label='main data')
plt.plot(x,intercept + slope * x,'r',label='Line formed')
plt.legend()
plt.show()
```

```
import pandas as pd
data = pd.read_csv('headbrain.csv')
data.head()
x = data['Head Size'].values.reshape(-1,1)
y = data['Brain Weight']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred =  lr.predict(x_test)
df = pd.DataFrame({'Actual Values':y_test,'Predicted vlaues':y_pred})
from sklearn.metrics import mean_squared_error,r2_score
print('Mean Squarred Error ',np.sqrt(mean_squared_error(y_test,y_pred)))

print('r2 score', r2_score(y_test,y_pred))
print(lr.intercept_)
print(lr.coef_)
import matplotlib.pyplot as plt
plt.plot(x,y,'o')
plt.show()
```
